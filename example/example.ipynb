```python
import sys
sys.path.append("../")
from bounding_box import BoundingBox
import matplotlib.pyplot as plt
import cv2
```

```python
bb = BoundingBox([747.31, 41.473, 1140.4, 712.92])
bb2 = BoundingBox((700, 20, 900, 350))
bb3 = BoundingBox((200, 200, 400, 400))

line1 = (100, 250, 350, 250)
line2 = (100, 100, 350, 100)

img = cv2.imread("./zidane.jpg")

img = cv2.rectangle(img, bb[:2], bb[2:], (255,0,0), 2)
img = cv2.rectangle(img, bb2[:2], bb2[2:], (0,255,0), 2)
img = cv2.rectangle(img, bb3[:2], bb3[2:], (0,0,255), 2)

img = cv2.line(img, (100,250), (350, 250), (255,255,0), 2)
img = cv2.line(img, (100,100), (350, 100), (0,255,255), 2)

plt.imshow(img[:,:,::-1])
plt.plot()
```

```python
print("Blue box is intercepting green box: ", bb.box_intercept_box(bb2.bounding_box))
print("Blue box is intercepting red box: ", bb.box_intercept_box(bb3.bounding_box))
print("Red box is intercepting blue line: ", bb3.box_intercept_line(line1))
print("Red box is intercepting yellow line: ", bb3.box_intercept_line(line2))
```

```python
img = cv2.imread("./zidane.jpg")

# Plot normal box
img = cv2.rectangle(img, bb[:2], bb[2:], (255,0,0), 2)

# Change box size to 10% bigger
bb.change_size(1.1)

# Plot box that now is 10% bigger
img = cv2.rectangle(img, bb[:2], bb[2:], (0,255,0), 2)

# Create another box from bb 50% smaller and preserving bb size 
bb4 = bb.change_size(0.5, False)

# Plot box originated from bb
img = cv2.rectangle(img, bb4[:2], bb4[2:], (0,0,255), 2)

plt.imshow(img[:,:,::-1])
plt.plot()
```

```python
print("How much of green box is in red box: ", bb.iou(bb4.bounding_box))
print("How much of red box is in green box: ", bb4.iou(bb.bounding_box))
```

```python
print("Bounding box values: ", bb4.bounding_box)
print("Bounding box listed values: ", bb4.list_bounding_box)
print("Bounding box middle: ", bb4.middle)
print("Bounding box listed middle: ", bb4.list_middle)
print("Bounding box dimensions: ", bb4.dimensions)
print("Bounding box listed dimensions: ", bb4.list_dimensions)
print("Bounding box area: ", bb4.area)
print("Bouding box walls coordinates: ", bb4.walls)
```
